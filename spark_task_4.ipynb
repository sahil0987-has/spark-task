```python
df4 = df.loc[:,'attacktype1':'attacktype1_txt']
df5 = pd.concat([df4, df3], axis=1)
```

```python
typeKill = df5.pivot_table(columns='attacktype1_txt', values='nkill', aggfunc='sum')
typeKill 
```

```python
typeKill.info()
```

```python
labels = typeKill.columns.tolist()  
transpoze = typeKill.T  
values = transpoze.values.tolist()
fig, ax = plt.subplots(figsize=(15,15), subplot_kw=dict(aspect="equal"))
plt.pie(values, startangle=90, autopct='%.2f%%')
plt.title('Types of terrorist attacks')
plt.legend(labels, loc='upper right', bbox_to_anchor = (1.3, 0.9), fontsize=15) 
plt.show()
```
